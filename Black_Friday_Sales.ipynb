```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
df.head(10)
```

```python
gender_map={'F':0,'M':1}
df['Gender']=df['Gender'].replace(gender_map)
df.head(10)
```

```python
df['stay_city'] = df['stay_city'].str.replace('+','')
df['stay_city'] = df['stay_city'].astype(int)
# converted to int
df.head(10)
```

```python
df['Product2'] = df['Product2'].fillna(0)
df
```

```python
df['Product3'] = df['Product3'].fillna(0)
df
```

```python
df1=df.groupby('Age')['Purchase'].mean().sort_values(ascending= False)
df1
plt.figure(figsize=(8,4))
sns.barplot(x=df1.index , y= df1.values)
plt.title("Age Vs Purchased Barplot")
plt.show()
```

```python
df2=df.groupby('Occupation')['Purchase'].mean().sort_values(ascending =False)
df2
plt.figure(figsize=(12,4))
sns.barplot(x=df2.index , y=df2.values)
plt.title('Occupation Vs Purchased Barplot')
plt.show()
```
